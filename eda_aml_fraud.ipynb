```python
import pandas as pd
df=pd.read_excel('transactions.xlsx')
```

```python
df.head(2)
df.isnull().sum()
df.duplicated().sum()
df.info()
df.describe()
df["is_fraud"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.boxplot(x="is_fraud", y="amount", data=df)
plt.xlabel("Suspicious Activity (0 = No, 1 = Yes)")
plt.ylabel("Transaction Amount")
plt.title("Transaction Amount vs Suspicious Activity")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="is_fraud", y="transactions_last_24h", data=df)
plt.xlabel("Suspicious Activity (0 = No, 1 = Yes)")
plt.ylabel("Transactions in Last 24 Hours")
plt.title("Transaction Velocity vs Suspicious Activity")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    x="customer_risk_rating",
    y="transactions_last_24h",
    hue="is_fraud",
    data=df
)
plt.xlabel("Customer Risk Rating")
plt.ylabel("Transactions in Last 24 Hours")
plt.title("Transaction Velocity by Customer Risk Rating and Suspicious Activity")
plt.legend(title="Suspicious Activity", labels=["No", "Yes"])
plt.show()
```
